# MLP Example

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import hp_tuning
from mlp import MLP
```

## Create A Regression Dataset

```python
T = 3
X = np.linspace(-T, T, 21)
y = np.sin(np.pi*X/4)
```

```python
plt.figure()
plt.plot(X, y, 'rx')
plt.title("Input Dataset")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Hyper-parameter Tuning

```python
n1 = [5, 10, 100, 1000]
lr1 = [0.0001, 0.00001]

hp_tuning(
    X=X,
    y=y,
    MLP=MLP,
    neurons=n1,
    learning_rate=lr1,
    iterations=1000,
)
```

```python
# 5 neurons is so close to 100
n2 = [5, 100]
lr2 = [0.001, 0.0001, 0.00001]

hp_tuning(
    X=X,
    y=y,
    MLP=MLP,
    neurons=n2,
    learning_rate=lr2,
    iterations=750,
)
```

```python
n3 = [5, 100]
lr3 = [0.001, 0.0005]

hp_tuning(
    X=X,
    y=y,
    MLP=MLP,
    neurons=n3,
    learning_rate=lr3,
    iterations=750,
)
```

```python
n4 = [50, 100, 150]
lr4 = [0.0005]

hp_tuning(
    X=X,
    y=y,
    MLP=MLP,
    neurons=n4,
    learning_rate=lr4,
    iterations=750,
)
```

```python
n5 = [100]
lr5 = [0.0005, 0.0008]

hp_tuning(
    X=X,
    y=y,
    MLP=MLP,
    neurons=n5,
    learning_rate=lr5,
    iterations=750,
)
```

```python
n6 = [100]
lr6 = [0.0008]

hp_tuning(
    X=X,
    y=y,
    MLP=MLP,
    neurons=n6,
    learning_rate=lr6,
    iterations=2000,
    n_fit=1,
)
```

## Create The Final Network

```python
# Tuned hyperparameters
n = 100
lr = 0.0008
iterations = 1200

# Create the network
mlp = MLP(n_features=1, hidden_neurons=n)

# Train the network
mlp.fit(X, y, iterations=iterations, learning_rate=lr, return_error=False)
```

## Evaluate The Network

```python
# Infere with the network
X_test = np.linspace(-3, 3, 500)
predictions = mlp.predict(X_test)
```

```python
plt.figure()
plt.plot(X, y, 'r.')
plt.plot(X_test, predictions, 'b')
plt.title("Regression Result")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["train", "test"])
plt.show()
```
